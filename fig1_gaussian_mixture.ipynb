# Mixture Gaussian Plot
This notebook is to plot the figures of centered Gaussian mixture and non-centered Gaussian mixture to show that centered Gaussian mixture distribution is more "Gaussian-like" (Fig. 1).

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set_style("ticks")
sns.set_context(
    "paper",
    font_scale=2.0,
    rc={
        "lines.linewidth": 2.5
    }
)
```

```python
def gauss(x, mu, sigma):
    return np.exp(- (x - mu)**2 / (2 * sigma**2)) / np.sqrt(2 * np.pi * sigma**2)

np.random.seed(10)

x = np.arange(-50, 50, step=0.01)

mu_1 = 0
mu_2 = 10

sigma_1 = np.sqrt(20)
sigma_2 = np.sqrt(50)

# Centered Gaussian Mixture
y_1_1 = gauss(x, mu_1, sigma_1)
y_1_2 = gauss(x, mu_1, sigma_2)
y_1 = y_1_1 + y_1_2

# Noncentered Gaussian Mixture
y_2_1 = gauss(x, mu_2, sigma_1)
y_2_2 = gauss(x, - mu_2, sigma_1)
y_2 = y_2_1 + y_2_2

# Plot the centered-Gaussian mixture
plt.plot(x, y_1, color=(0.372, 0.537, 0.537))
plt.plot(x, y_1_1, color=(0.372, 0.537, 0.537), alpha=0.2)
plt.plot(x, y_1_2, color=(0.372, 0.537, 0.537), alpha=0.2)

plt.xticks(color="None")
plt.yticks(color="None")
plt.tick_params(length=0)

plt.savefig('figs/gauss_mix_centered.pdf')

plt.show()

# Plot the noncentered-Gaussian mixture
plt.plot(x, y_2, color=(0.627, 0.352, 0.470))
plt.plot(x, y_2_1, color=(0.627, 0.352, 0.470), alpha=0.2)
plt.plot(x, y_2_2, color=(0.627, 0.352, 0.470), alpha=0.2)

plt.xticks(color="None")
plt.yticks(color="None")
plt.tick_params(length=0)

plt.savefig('figs/gauss_mix_noncentered.pdf')

plt.show()
```
